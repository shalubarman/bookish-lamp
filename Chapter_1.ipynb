```python
# https://pandas.pydata.org/docs/user_guide/10min.html
import pandas as pd
import numpy as np
```

```python
# create a series & dataframe
s = pd.Series([1,2,3,np.nan,5])
print(s)
df=pd.DataFrame(np.random.randn(3,4))
print(df)
```

```python
# DF properties & viewing data
df2 = pd.DataFrame({
    "A" :1.0,
    "B" :pd.Timestamp("20130201"),
    "C" :pd.Series(1, index=list(range(4)), dtype="float32"),
    "D" :np.array([3]*4, dtype="int32"),
    "E" :pd.Categorical(["test", "train", "test", "train"]),
    "F" :"col",
})
print(df2)
print("\n")
print("*"*30)
print("DF2 Datatypes")
print("*"*30)
print(df2.dtypes)
print("\n")
print("*"*30)
print("DF2 first 2 rows")
print("*"*30)
print(df2.head(2))
print("\n")
print("*"*30)
print("DF2 last 2 rows")
print("*"*30)
print(df2.tail(2))
print("\n")
print("*"*30)
print("DF2 Index")
print("*"*30)
print(df2.index)
print("\n")
print("*"*30)
print("DF2 column names")
print("*"*30)
print(df2.columns)
print("\n")
print("*"*30)
print("DF2 describe")
print("*"*30)
print(df2.describe())
print("\n")
print("*"*30)
print("DF2 Sort by Index and Values")
print("*"*30)
print(df2.sort_index(axis=1,ascending=False))
print(df2.sort_values(by="E"))
```

```python
# Selection
df2=pd.DataFrame({
    "AB" : pd.date_range("20220301", periods=5),
    "BC" : pd.Series(list(range(5))),
    "CD" : pd.Categorical(['Shxxx','Hmxxx','Jixxx','Smxxx','Lmxxx'])
})
# df2 = df2.set_index("AB")
print(df2)
print("\n")
print("*"*30)
print(df2.BC) # Equivalent to df2['BC']
print("\n")
print("*"*30)
print(df2[2:4])
# Selection by label
print("\n")
print("*"*30)
print(df2.loc[:,['BC']])
```
